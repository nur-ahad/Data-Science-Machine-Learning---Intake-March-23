```python
# Finding correlation among features (with numerical variable)
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Finding highly correlated features

high_corr = corr_matrix[abs(corr_matrix) > 0.8]
print(high_corr)
```

```python
# Finding moderately correlated features

mod_corr = corr_matrix[abs(corr_matrix) > 0.4]
print(mod_corr)
```

```python
#Finding low correlation features

low_corr = corr_matrix[abs(corr_matrix) > 0.2]
print(low_corr)
```
